```python
# the output csv
school_dis21_dir = '../distances_and_durations/school_old/school_distance_duration(final).csv' 
```

```python
# add them together
df = pd.concat([sch1,sch2])
df = pd.concat([df, sch3])
df = pd.concat([df, sch4])
df = pd.concat([df, sch5])
df = pd.concat([df, sch6])
df = pd.concat([df, sch7])
df = pd.concat([df, sch8])
df = pd.concat([df, sch9])
df = pd.concat([df, sch10])
df = pd.concat([df, sch11])
df = pd.concat([df, sch12])
df = pd.concat([df, sch13])
df = pd.concat([df, sch14])
df = pd.concat([df, sch15])
df = pd.concat([df, sch16])
df = pd.concat([df, sch17])
df = pd.concat([df, sch18]).reset_index(drop=True)

```

```python
df
```

```python
# add rantal house's postcode and price
rent = pd.read_csv("../rent_data/renew(final).csv")
new_data = df
postcode = []
for i in range(len(new_data)):
    postcode.append(new_data['rent_address'][i][-13:-9])
new_data['postcode'] = postcode
price =[]
for rent_a in new_data['rent_address']:
    price.append(rent[rent['listingSlug'] == rent_a]['price'].tolist()[0])
new_data['price'] = price
new_data = new_data.sort_values(by=["postcode"],ascending=True).reset_index(drop=True)
```

```python
rent= pd.read_csv("/home/ads/generic-real-estate-consulting-project-group-55/rent_data/renew(final).csv", index_col=0)
rent_c = pd.read_csv("/home/ads/generic-real-estate-consulting-project-group-55/rent_data/rent_cleaned.csv", index_col=0)
slice_lable = (
    rent[['listingSlug']].apply(tuple, axis=1)
    .isin(rent_c[['listingSlug']].apply(tuple, axis=1)
          .to_list()
         )
)

rent[~slice_lable]
```

```python
index = []
for i in range(len(rent[~slice_lable]['listingSlug'])):
    index.append(new_data[new_data['rent_address'] == rent[~slice_lable]['listingSlug'].tolist()[i]].index.tolist()[0])
new_data = new_data.drop(index = index, axis=0)
```

```python
new_data.to_csv(school_dis21_dir)
```
