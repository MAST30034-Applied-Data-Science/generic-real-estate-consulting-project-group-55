<font size=3> Population Data

```python
population_df=pd.read_excel('../data/raw/population.xlsx',sheet_name='Table 3')
population = population_df[8:]
population.to_csv('../data/raw/population_2.csv', encoding='utf-8')
population
```

```python
population=pd.read_csv('../data/raw/population_2.csv',header=0,names=['Year','S/T code','S/T name', 'GCCSA code', 'GCCSA name', \
                     'SA4 code', 'SA4 name', 'SA3 code', 'SA3 name', 'SA2 code', \
                     'SA2 name', '0-4','5-9', '10-14','15-19', '20-24','25-29',\
                    '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74',\
                    '75-79','80-84','85 and over', 'total persons'])
VIC_population = population.loc[population['S/T name']=='Victoria']
```

```python
select_population = VIC_population.loc[:, ['Year','S/T name', 'SA2 name', '0-4','5-9', '10-14','15-19', '20-24','25-29',\
                    '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74',\
                    '75-79','80-84','85 and over', 'total persons']] 
```

```python
pop_18to21 = select_population.loc[(select_population['Year'] == '2018') | (select_population['Year']== '2019') | (select_population['Year']=='2020') | (select_population['Year']=='2021')] 
pop_18to21



```

```python
postcode_SA2 = pd.read_csv('../data/raw/postcode_info.csv')
postcode_SA2 
```

```python
attribute = postcode_SA2.loc[:,['Postcode','Longitude','Latitude','SA2 NAME 2016']]
drop_dup = attribute.drop_duplicates(subset=['Postcode'])
drop_dup = drop_dup.drop_duplicates(subset=['Longitude','Latitude'])
rename = drop_dup.rename(columns={'SA2 NAME 2016':'SA2 name'})
```

```python
rename
```

```python
merge_data = pd.DataFrame(pd.merge(rename, pop_18to21, on='SA2 name', how='inner'))
```

```python
merge_data.to_csv('../data/curated/population.csv')
```
